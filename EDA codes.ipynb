KOLOM faktor_beli

```python

# Daftar pilihan yang valid
valid_choices = ["rasa", "harga", "kualitas bahan baku", "kenyamanan lokasi", "aroma"]

# Memeriksa jawaban yang berada di luar pilihan yang valid
invalid_choices = df[~df['faktor_beli'].isin(valid_choices)]

# Menampilkan jawaban yang tidak valid
print(invalid_choices['faktor_beli'].unique())
```

```python
# Fungsi untuk merubah nilai-nilai pada kolom faktor_beli
def standardize_factor_beli(value):
    value = value.lower()  # Mengubah nilai menjadi huruf kecil
    if value in valid_choices:
        return value
    return 'other'
```

```python
# Menerapkan fungsi pada kolom faktor_beli
df['faktor_beli'] = df['faktor_beli'].apply(standardize_factor_beli)

# Menampilkan hasil perubahan
print(df['faktor_beli'].unique())
```

```python
def cronbach_alpha(df):
    #jumlah item (kolom)
    k = df.shape[1]
    
    #item variances
    item_variances = df.var(axis=0, ddof=1)
    
    #total variance
    total_variance = df.sum(axis=1).var(ddof=1)
    
    #hitung alpha Cronbach
    alpha = (k / (k - 1)) * (1 - item_variances.sum() / total_variance)
    
    return alpha

# Hitung alpha Cronbach untuk DataFrame yang telah difilter
alpha = cronbach_alpha(df_rel)
print(f"Cronbach's Alpha: {alpha}")
```

```python
# Hitung mean observasi untuk variabel 'jumlah_kopi'
mean_observed = df['jumlah_kopi'].mean()
print(f"Mean Observed: {mean_observed}")

# Nilai error yang diketahui
error = 0.1
print(f"Error: {error}")

# Hitung true value menggunakan rumus μ_i = Y_it - ε_it
true_value = mean_observed - error
print(f"True Value (μ_i): {true_value}")
```

```python
# Hitung bias
bias = sum((df['jumlah_kopi'] - true_value) / len(df))

print(f"Bias: {bias}")
```
